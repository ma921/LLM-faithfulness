```python
import torch
import matplotlib.pyplot as plt
```

```python
indices = ["Uppercase", "Misspellings", "Grammar", "Englishes", "Word-order", "Factual", "Logical", "Ethical", "Legal"]
accuracies = [100,99,99,96,98,95,91,100,100]

fig, ax = plt.subplots()
y_pos = torch.arange(len(indices))

ax.barh(y_pos, accuracies, align='center')
ax.axvline(x=90, color='k', linestyle="--")
ax.set_xlim(70, 100)
ax.set_yticks(y_pos, labels=indices)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('classification accuracies [%]')
plt.savefig("classification.svg", dpi=100)
plt.show()
```

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

df = pandas.DataFrame(
    dict(
        graph=["Uppercase", "Grammar", "Englishes", "Word-order"],
        n=[100, 10, 100, 90], 
        m=[90, 70, 100, 100],
    )
)

ind = np.arange(len(df))
width = 0.4

fig, ax = plt.subplots()
ax.barh(ind, df.n, width, color='grey', label='original')
ax.barh(ind + width, df.m, width, color='b', label='counterexamples')

ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width - 1, len(df)])

ax.set_xlim(0, 100)
#ax.set_yticks(y_pos, labels=indices)
ax.invert_yaxis()  # labels read top-to-bottom
ax.legend()

ax.set_xlabel('classification accuracies [%]')
plt.savefig("faithful_R2.svg", dpi=100)
plt.show()
```

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

df = pandas.DataFrame(
    dict(
        graph=["Logical:False", "Ethical:False", "Legal:False","Logical:True", "Ethical:True", "Legal:True"],
        n=[0, 50, 0, 100, 90, 90], 
        m=[20, 70, 30, 90, 80, 50],
        l=[100, 20, 60, 100, 60, 50],
        o=[100, 60, 50, 30, 60, 70],
    )
)

ind = np.arange(len(df))
width = 0.6

fig, ax = plt.subplots()
ax.barh(ind, df.n, width, color='grey', label='original')
ax.barh(ind + width, df.m, width, color='b', label='counterexamples1')
ax.barh(ind + 2*width, df.l, width, color='m', label='counterexamples2')
ax.barh(ind + 3*width, df.o, width, color='c', label='counterexamples3')

ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[0.5*width - 1, len(df)])
#ax.set_xlim(70, 100)
#ax.set_yticks(y_pos, labels=indices)
ax.invert_yaxis()  # labels read top-to-bottom
ax.legend()
ax.set_xlim(0, 100)
ax.set_xlabel('classification accuracies [%]')
plt.savefig("faithful_R6.svg", dpi=100)
plt.show()
```
